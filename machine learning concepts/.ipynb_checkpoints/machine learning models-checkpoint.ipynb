### Improving performance   
 
After evaluating the performance we need to improve the model there are several ways we can do it:
<ul>
    <li>Dimensionality reduction</li>
    <li>Hyperparameter tuning</li> 
    <li>Ensemble methods</li>
</ul>


<b>Dimensionality reduction</b>

<ul>
<li>A dimension denotes the number of features in your data, so dimensionality reduction essentially means reducing the number of features</li>  
<li>Irrelevance: some features don't carry useful information</li> 
<li>Correlation: some features carry similar information. We can use only one feature or collapse multiple features into one underlaying feature</li>
</ul>
<b>Hyperparameter tuning</b>
<ul>
    <li>A hyperparameter is a model parameter that is chosen by you before training begins. (This is in contrast to parameters, which are determined by the model training.) The hyperparameters available to set differ between types of model.</li>
    <li>SVM algorithm hyperparameters: kernel, C, degree, gamma, shrinking, coef0, tol...</li>
</ul>
<b>Ensemble methods</b>
<ul>
    <li>Technique that combines several models in order to produce one optimal model</li>
</ul>
<div>
    
</div>
<ul>
    <li>Ensemble methods classification - Take the most common predictions amoung the models</li>
    <li>Ensemble methods regression - Average of the all the models output</li>
</ul>
