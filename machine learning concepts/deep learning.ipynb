# Deep learning  

<ul>
<li>Deep learning uses an algorithm called  neural networks, which are loosely inspired by biological neural networks in human brains</li>
<li>Neurons, also sometimes called nodes, are the basic unit of neural networks </li>
<li>Deep learning is a special type of machine learning that can solve more complex problems but it requires much more data than traditional machine learning</li>
<li>It is best used in cases where the inputs are less structured, such as large amounts of text or images.</li>
</ul>
<br>
Lets see how it works  
<br>
Imagine you want to predict a box office revenue for an upcomming movie

</div>

    <li>In neural network this model will represent like the node next to the graph</li>
    <li>If we have more features we can train models and we can connect those node to neural network and get the predicting box office revenue</li>
</ul>
<div>
    
</div>
<ul>
    <li>Its job is to map relationships between different combinations of variables to the desired output</li>
</ul>    
<br>
<b>When to use deep learning</b>
<ul>
    <li>Deep learniing can outperform other techniques if the data size is large</li>
    <li>When there is a lack of domain knowledge for understanding the features</li>
    <li>Complex problems like computer vision, NLP etc.</li>
</ul>

### Natural Language Processing  
<ul>
    <li> The ability for computers to understand the meaning of human language. </li>
    <li> <b>Bag of Words</b> (BOW) is a method to extract features from text documents.</li>
</ul>

<br>
Limitations of BOW
<br>
<ul>
    <li>Word count don't help us consider synonums</li>
    <li>Want to group this as a single feature</li>
    <li>One way to solve this <b>Word embaddings</b> - Create features that group similar words </li>
</ul>
